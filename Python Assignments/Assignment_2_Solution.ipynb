### 1. Write a Python program to convert kilometers to miles?
##### Ans:.

```python
km = int(input("Enter Kilometers = "))
miles = km/1.60934
print("After conversion distance in miles = ", miles)
```

### 2.Write a Python program to convert Celsius to Fahrenheit?
##### Ans:-

```python
c=float(input("Enter Celsius = "))
f = (9*c/5)+32
print("After conversion fahrenheit = ", f)
```

### 5. Write a Python program to swap two variables without temp variable?
##### Ans:-

```python
var1 = 6
var2 = 4
print('Before swap:\nvar1 = {} and var2 = {}'.format(var1, var2))
var2 = var1 + var2
var1 = var2 - var1
var2 = var2 - var1
print('\nAfter swap:\nvar1 = {} and var2 = {}'.format(var1, var2))
```
